```python
## using python standard counter to increment and read data like this
from collections import Counter
with open('data.csv') as csv_file: ## using standard lib to read csv
    csv_reader = csv.DictReader(csv_file) ## looping and reading data into a dict

    lang_count = Counter() ## instantiating our counter

    for row in csv_reader:
        lang_count.update(row['LanguagesWorkedWith'].split(';')) #updating our counter instance on every loop

print(lang_count)
```

```python
print(lang_count.most_common(10))##most common method to get top values
```

```python
lang = [] ## setting empty lists to loop into and break out the tuple of most common lang and count
popular = []

for item in lang_count.most_common(10):
    lang.append(item[0])
    popular.append(item[1])

print(lang,popular)
```

```python
## we can do the above in one line with the zip function
plt.barh(lang,popular) ## barh for a horizontal

plt.title('Most pop langs') 
plt.xlabel('Count of ppl')

plt.tight_layout() 


plt.show()
```

```python

lang.reverse()## reversing our lists to see most pop on top
popular.reverse()

plt.barh(lang,popular) ## barh for a horizontal

plt.title('Most pop langs') 
plt.xlabel('Count of ppl')

plt.tight_layout() 


plt.show()
```

```python
## using pandas to load in the data
import pandas as pd

df = pd.read_csv('data.csv')
ids=df['Responder_id']
languages = df['LanguagesWorkedWith']

lang_count2 = Counter()
for lang in languages:
        lang_count2.update(lang.split(';')) 

print(lang_count2)
```

```python
lang = [] ## setting empty lists to loop into and break out the tuple of most common lang and count
popular = []

for item in lang_count2.most_common(10):
    lang.append(item[0])
    popular.append(item[1])


lang.reverse()## reversing our lists to see most pop on top
popular.reverse()

plt.barh(lang,popular) ## barh for a horizontal

plt.title('Most pop langs') 
plt.xlabel('Count of ppl')

plt.tight_layout() 


plt.show() ## with pd still works as before
```
