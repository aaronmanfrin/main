```python
## dropping missing data such as np.nan, None, 'NA'
df.dropna() ## dropna method will drop all rows with missing data
```

```python
df.dropna(axis='index',how='any') ##axis will be set to index or columns, any will drop rows with any missing values
```

```python
df.dropna(axis='index',how='all') ## all will drop anything that is missing all values
```

```python
## dropping values with missing data in specific locations
df.dropna(subset=['EdLevel','YearsCodePro'],how='any') ##subset will define which specific columns to look at

```
